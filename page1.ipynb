```python
def extract(str):
    df_str = df.loc[df.iloc[:, 0] == str]
    list_str = df_str.iloc[:,1].tolist()
    return list_str
```

```python
list_id = extract("ID")
list_kaishamei = extract("会社名")
list_bushou1 = extract("部署・役職")
list_bushou2 = extract("部署・役職区分")
list_shimei = extract("氏名")
list_gyoushu = extract("業種区分")
list_shougyouinn = extract("従業員数区分")
list_todofukenn = extract("都道府県")
list_homepage = extract("会社ホームページURL")
list_mokuteki = extract("利用目的")
#list_yousann = extract("導入予算額")
#list_AI = extract("【ＡＩ・業務自動化 関連】")
list_kyoumi = extract("興味のある製品・ソリューション")
list_kaidai = extract("解決したい課題")
list_kannkyou = extract("解決したい課題、現状のシステム環境")
list_youteibi = extract("来場予定日")
data_tuples = list(zip(list_id,list_kaishamei,list_bushou1,list_bushou2, 
                       list_shimei, list_gyoushu, list_shougyouinn,list_todofukenn, 
                       list_homepage, list_mokuteki,
                       list_kyoumi, list_kaidai, list_kannkyou, list_youteibi
                      ))
df_new = pd.DataFrame(data_tuples, columns = title)
df_new
```
